```python
import numpy as np 
import matplotlib.pyplot as plt 
```

Functions to generate sin(x) and cos(x) 

```python

#generating x values 
x_arr = np.linspace(0, 1, 100) 

#function that return numpy sin 
def my_sin(x):
    return np.sin(x) 

#function that returns numpy cosine
def my_cos(x): 
    return np.cos(x) 
```

Functions to generate plots of sin(x) and cos(x) 

```python
#ax is the array of axes 
f, axs = plt.subplots(1, 2)

#sin panel 
axs[0].plot(x_arr, my_sin(x_arr))
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[0].set_title(r'$\sin(x)')

#cos panel 
axs[1].plot(x_arr, my_cos(x_arr))
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')
axs[1].set_title(r'$\cos(x)$')

f.subplots_adjust(wspace = 0.4) 
f.savefig('sin_and_cos.pdf')
```
